# Task 2: Clean and Preprocess data

## Task 2.1: Merge data of all 12 months

```python
frames = []
all_length = []
for file in os.listdir(path):
    if file.endswith('.csv'):
        filepath = path + file
        df1 = pd.read_csv(filepath)
        frames.append(df1)
        result = pd.concat(frames)
        length_1month = len(df1.index)
        all_length.append(length_1month)
df = result
```

```python
df
```

## Task 2.2: Add 'Month' column

```python
df['Month'] = df['Order Date'].str[0:2]
df.head(15)
```

```python
print(set(df['Month']))
```

## Task 2.3: Get rid of 'Nan' and 'Or' value

```python
df = df.dropna(how = 'all')
df.head()
```

```python
df = df[df['Month'] != 'Or']
df.head()
```

# Task 3: Reporting

## Task 3.1: What was the best months for sales? How much was earned that month?

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'],downcast = 'float')
df['Price Each'] = pd.to_numeric(df['Price Each'],downcast = 'float')
```

```python
print(df['Quantity Ordered'].dtypes)
print(df['Price Each'].dtypes)
```

```python
df['Sales'] = df['Quantity Ordered']*df['Price Each']
df.head()
```

```python
moved_column = df.pop('Sales')
df.insert(4,'Sales',moved_column)
df
```

```python
df.groupby('Month').sum()['Sales']
```

```python
sales_value = df.groupby('Month').sum()['Sales']
sales_value.max()
```

```python
months = range(1,13)
plt.bar(x=months, height = sales_value)
plt.xticks(months)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

## Task 3.2: What city has the best sales?

```python
df.head()
```

```python
address_to_city = lambda address:address.split(',')[1]
```

```python
df['City'] = df['Purchase Address'].apply(address_to_city)
df.head()
```

```python
df.groupby('City').sum()['Sales']
```

```python
sales_value_city = df.groupby('City').sum()['Sales']
sales_value_city.max()
```

```python
cities = [city for city, sales in sales_value_city.items()]
```

```python
plt.bar(x=cities, height = sales_value_city)
plt.xticks(cities, rotation = 90, size = 8)
plt.xlabel('Cities')
plt.ylabel('Sales')
plt.show()
```

## Task 3.3: What time should we display ads to maximize the likelihood of customer's buying product?

```python
df.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
sample_date = '01/22/19 21:25'
sample_date_to_datetime = pd.to_datetime(sample_date)
sample_date_to_datetime.hour
```

```python
df['Hours'] = df['Order Date'].dt.hour
```

```python
df.head()
```

```python
sales_value_hours = df.groupby('Hours').count()['Sales']

hours = [hour for hour, sales in sales_value_hours.items()]

plt.plot(hours,sales_value_hours)
plt.grid()
plt.xticks(hours, size = 8)
plt.xlabel('Hours')
plt.ylabel('Sales')
plt.show();
```
